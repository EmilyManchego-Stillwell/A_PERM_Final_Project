```python
import pandas as pd
```

```python
fires = pd.read_csv("WFIGS_-_Wildland_Fire_Locations_Full_History.csv")

fires.head()
```

```python
fires["YEAR"]=pd.DatetimeIndex(fires["FireDiscoveryDateTime"]).year
small_fires = fires[["YEAR", "POOCounty", "X"]]
```

```python
# group by year, then county and count the X column so that we get a count of the number of wildfires in each county in each year
final = small_fires.groupby(by=["YEAR", "POOCounty"]).count()

final
```

```python
# save grouped df as a csv
final.to_csv("wildfire_by_county.csv")
```

```python
# save the df as a dictionary 
county_year_dict = final.to_dict()['X']

county_year_dict
```

```python
# test to see if try/except will work by calling a (year, county) key from the dictionary
counties = ['Archuleta', 'Baca']
years = [2015, 2016, 2017, 2018]

for year in years:
    for county in counties:
        try:
            print(county_year_dict[(year, county)])
        except KeyError:
            print(0)
```
